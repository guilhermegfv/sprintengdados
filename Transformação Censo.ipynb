```
from pyspark.sql.functions import *
```

```
dbutils.fs.ls("dbfs:/dbfs/")
```

```
df = spark.read.format("delta").load("dbfs:/dbfs/censo")
```

```
df = df.filter(df['Variável'] == 'População residente')
df = df.select(['Estado','MunicípioCódigo','Cidade','Valor'])
df = df.withColumnRenamed('Valor','Habitantes')
```

```
df.repartition(20)\
              .write.format("delta")\
              .mode("overwrite")\
              .partitionBy('AnoMes')\
              .option("overwriteSchema", "true")\
              .save("/dbfs/censo_refinado")
```
